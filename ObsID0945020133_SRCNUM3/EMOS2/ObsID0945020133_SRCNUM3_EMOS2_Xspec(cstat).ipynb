```python
from xspec import *
from stingray import Lightcurve
from astropy.io import fits
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
import numpy as np

class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'

plt.rcParams['figure.figsize'] = [12,7]
```

### Redshift

```python
def redshift(d_Mpc):
    H0 = 70 # Hubble constant (km/s/Mpc)
    c = 299792.458 # km/s
    return H0*d_Mpc/c

z = redshift(7173.0551)
print("Redshift : z =", z)
```

## Fit statistic

```python
Fit.statMethod = "cstat"
Fit.nIterations = 100000
Fit.delta = 1e-4
Fit.criticalDelta = 1e-4
```

```python
AllData.clear()
spec = Spectrum(glob.glob('ObsID*_SRCNUM*_E*_spectrum_grp_cstat.fits')[0])
```

## Flux
### X-ray flux
- X-ray flux of the best-fitting model ``Apec`` with C-statistic = 112.6732 using 117 bins: **1.3409e-13 ergs/cm^2/s**.
- X-ray flux of the second best-fitting model ``Bremss`` with C-statistic = 112.7949 using 117 bins: **9.9688e-14 ergs/cm^2/s**.

```python
Flux_a = 1.3409e-13 # ergs/cm^2/s
Flux_b = 9.9688e-14 # ergs/cm^2/s
print(color.BOLD+"X-ray flux : "+color.END)
print(f"Apec : F_x-ray = {Flux_a} ergs/cm^2/s")
print(f"Bremss : F_x-ray = {Flux_b} ergs/cm^2/s")
```

### Optical flux
No GAIA information, so we use the lower limit of 20.7 mag.

```python
def F_optical(mag):
    F0 = 1.05*10**(-5) # GAIA zero point magnitude
    return F0*10**(-0.4*mag)
    
G = 20.7

F = F_optical(G)

print(color.BOLD+"Optical flux (lower limit) : "+color.END+f"F_optical = {F} ergs/cm^2/s")
```

### X-ray flux / Optical flux ratio

```python
print(color.BOLD+"X-ray flux / Optical flux ratio : "+color.END)
print(f"Apec : F_x-ray/F_optical = {Flux_a/F} ergs/cm^2/s")
print(f"Bremss : F_x-ray/F_optical = {Flux_b/F} ergs/cm^2/s")
```

## Luminosity

```python
"""GLADE"""

d_kpc = 7173055.1 # Distance en kpc
d_cm = d_kpc*3.0857*10**21 # Distance en cm, avec 1 pc = 3.0857*10^18 cm

# La source est supposée sphérique pour le calcul de la luminosité : L = 4*pi*F*D^2
L_apec = 4*np.pi*Flux_a*(d_cm)**2 # erg/s
L_bremss = 4*np.pi*Flux_b*(d_cm)**2 # erg/s

print(color.BOLD+"(GLADE)"+color.END)
print(f"Distance : d = {d_cm} cm")
print(f"Apec : L_x-ray = {L_apec} erg/s")
print(f"Bremss : L_x-ray = {L_bremss} erg/s")
```
